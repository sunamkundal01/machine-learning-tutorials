```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.rottentomatoes.com/m/my_old_ass/reviews'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
```

```python
print(soup)
```

```python
movie_name=soup.find('h2',{'data-qa':'reviews-title'}).text
print(movie_name)
```

```python
names=soup.find_all('a',class_='display-name')
print(names)
```

```python
reviews=soup.find_all('p',class_='review-text')
print(reviews)
```

```python
rating = soup.find_all('p', class_='original-score-and-url')
original_scores = []  # Initialize a list to store all scores

if rating:
    # Loop through each element in the ResultSet
    for item in rating:

        # Extract the text that includes "Original Score"
        score_text = item.get_text(strip=True)

        # Extract only the score part (e.g., "3/5") using string manipulation
        original_score = score_text.split("Original Score:")[-1].split("|")[0].strip()
        # The if statement needs an indented block to be executed when the condition is true.
        if item.get_text(strip=True).startswith("Full Review"):
            # Append the extracted score to the list if the condition is True.
            original_scores.append(original_score)

rating = original_scores  # Assign the list of scores to rating
print(rating)
```

```python
date=soup.find_all('span', {'data-qa': 'review-date'})
print(date)
```

```python
review_table = [(date_item.text.strip(), movie_name, name.text.strip(), rating_item, review_item.text.strip())
                for date_item, _, name, rating_item, review_item in zip(date, movie_name, names, rating, reviews)]  # Use movie_name_list instead of movie_name, and ignore the string in the loop
print(review_table)
```

```python
data=pd.DataFrame(review_table,columns=['date','Movie','Name','rating','Review'])
print(data)
```

```python
df = pd.DataFrame(data)

# Save to Excel
df.to_excel('output.xlsx', index=False)
```

```python
import nltk
import string
from nltk.corpus import stopwords
import pandas as pd

# Download the stopwords
nltk.download('stopwords')

# Define the function for text processing
def preprocess_text(text):
    # Tokenization: Split sentences into words
    tokens = text.split()

    # Lowercase: Convert everything to lowercase
    tokens = [word.lower() for word in tokens]

    # Remove Punctuation: Strip out punctuation marks
    tokens = [word.strip(string.punctuation) for word in tokens]

    # Remove Stop Words: Filter out common words that don’t add much meaning
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    return tokens

# Check the actual column names in your DataFrame
print(data.columns)

# Replace 'Review Text' with the actual column name
# For example, if the actual column name is 'review', use:
# data['processed_text'] = data['review'].apply(preprocess_text)
# Assuming the actual column name is 'Review':
data['processed_text'] = data['Review'].apply(preprocess_text)

# Display the processed text
print(data[['Review', 'processed_text']].head()) # Also change here to reflect the actual column name
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Assuming you already have a 'processed_text' column with the cleaned text
# You need to convert lists of tokens back to a format that CountVectorizer can use
# Typically, you would join the tokens back into a single string per document
data['cleaned_text'] = data['processed_text'].apply(lambda x: ' '.join(x))

# Create the Bag of Words model
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data['cleaned_text'])

# Convert to an array if needed
bow_array = X.toarray()

# Display the BoW array and the feature names
print(bow_array)
print(vectorizer.get_feature_names_out())
```
