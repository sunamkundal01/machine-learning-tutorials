**Bag of Words**


A bag of words is a repersentaion of text taht describes the occurances of words with in a document, It involves two things :
1.vocablary of words
2. measure of the presence of known words.


```python
text_data=["The cat sat on the mat",
    "The dog sat on the log",
    "The cat chased the dog",]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer(stop_words='english')
```

```python
#fit the data
bow.fit(text_data)
```

```python
#get the volcabulary List
bow.get_feature_names_out()
```

```python
# tranform the data
bow_features=bow.transform(text_data)
```

```python
bow_features
```

```python
bow_features_array=bow_features.toarray()
bow_features_array
```

```python
print(bow.get_feature_names_out())
for sentense ,feature in zip(text_data,bow_features_array):
  print(sentense)
  print(feature)
```

# TF-IF(Term Freqency/ Inverse Documnet Frequency)
TF-IDF it is maeseure ,used in the field of information retrival and machine learning, that can quantify the importance or relevance of string repersentations(eords,phrases,lemmas,etc) in a documnet amongest a colection of documents.

```python
text_data=["The cat sat on the mat",
    "The dog sat on the log",
    "The cat chased the dog",]
```

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')
```

```python
# fit the data
tfidf.fit(text_data)
```

```python
#get the vocalabary list count how much times word occur
tfidf.vocabulary_
```

```python
tfidf_features=tfidf.transform(text_data)
tfidf_features
```

```python
tfidf_features_array=tfidf_features.toarray()
tfidf_features_array
```

```python
for sentense ,feature in zip(text_data,tfidf_features):
  print(sentense)
  print(feature)
```
